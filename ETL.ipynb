## 1. Find the average age of developers when they wrote their first line of code. 

```python
df["Age1stCode"]=df["Age1stCode"].astype(str)
def num(x):
    for word in x.split():
        if word. isdigit():
            return word
        else:
            return 0
       
df["Age1stCode"]=df["Age1stCode"].apply(num)
print("Average age is: ",df["Age1stCode"].astype(float).mean())
```

## 2. Deduce the percentage of developers who knew python in each country. 

```python
df["Python"]=df["LanguageWorkedWith"].astype(str).apply(lambda x:'python' in x.lower())
```

```python
df1=df[["Country","Python"]]
df2=df1[df1["Python"]==True]
df1=df1.groupby(["Country"]).count()
df2=df2.groupby(["Country"]).count()
```

```python
df3=df2/df1*100
df3
```

## 3. Generate the report for the average salary of developers based on the continent.


```python
# Add Continent Column 
df_continent=pd.read_csv("countryContinent.csv")
df_continent.head()
df = pd.merge(df, df_continent, how="left", on=["Country"])
# Check Null values
df["Continent"].isna().sum()
```

## 5 Generate the report for the people who code as a hobby based on their gender and continent. 

```python
list=["Man","Woman"]
def other(x):
    if x not in list:
        return "Other"
    else:
        return x
  
df["Gender"]=df["Gender"].apply(other)
df["Gender"].value_counts()
result=df[df["Hobbyist"]=='Yes']
```

```python
result2=df.groupby(["Continent","Gender"]).agg({'Hobbyist':'count'})
result1=result.groupby(["Continent","Gender"]).agg({'Hobbyist':'count'})
```

```python
Final_Result=(result1/result2)*100
# Percentage of Peopel with Coding as Hobby By continent and Gender
Final_Result
```

## 6. Generate the report for job and career satisfaction of developers based on their gender and continent.


```python
# Carrer Satisfaction
Car_sat=df.groupby(["Continent","Gender","CareerSat"]).agg(Careersatistified=('CareerSat','count'))
Total_1=df.groupby(["Continent","Gender"]).agg(Careersatistified=('CareerSat','count'))
Car_sat=(Car_sat/Total_1)*100
Car_sat
```

```python
# Job Satisfaction
Job_sat=df.groupby(["Continent","Gender","JobSat"]).agg(JobSatisfied=('JobSat','count'))
Total_2=df.groupby(["Continent","Gender"]).agg(JobSatisfied=('JobSat','count'))
Job_sat=(Job_sat/Total_2)*100
Job_sat
```

```python
# Job and Career satisfaction
pd.concat([Car_sat,Job_sat],axis=1)
```
